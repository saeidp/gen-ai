```python
from strands import Agent, tool
from strands.models import BedrockModel
from strands_tools import calculator
# model_id = "anthropic.claude-haiku-4-5-20251001-v1:0"
model_id = "global.anthropic.claude-haiku-4-5-20251001-v1:0"

model = BedrockModel(
    model_id=model_id,
)
agent = Agent(
    model=model,
    tools=[
        calculator
    ],
    system_prompt="""
    You're a helpful assistant. You can do simple math calculations,
    Always start by acknowledging the user's name
    """
)
```

```python
import json
payload = json.loads('{"prompt": "What is the 2 + 3"}')
user_input = payload.get("prompt")
response = agent(user_input)
print(response.message['content'][0]['text'] )
```

```python
import asyncio


async def stream_agent(prompt: str):
    final = None
    async for event in agent.stream_async(prompt):
        if event.get("data"):
            print(event["data"], end="", flush=True)
        if event.get("current_tool_use"):
            tool = event["current_tool_use"]
            print(f"\n[using tool: {tool.get('name')}]\n", end="")
        if event.get("result"):
            final = event["result"]
    print("\n\n--- complete ---")
    if final:
        print(final.message["content"][0]["text"])

await stream_agent("What is 2 + 3?")
```
